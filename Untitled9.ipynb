```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Data for percentage change in supply and time to sell
data = {
    'Property Type': ['Detached', 'Semi-Detached', 'Terraced', 'Flat'],
    'Supply Change (%)': [18, 29, 52, 30],
    'Time to Sell Change (%)': [11, 8, 14.8, 3.7]
}

# Create DataFrame
df = pd.DataFrame(data)

# Set up bar positions
x = np.arange(len(df))
bar_width = 0.35

# Define custom colors for each property type
supply_colors = ['skyblue', 'lightcyan', 'darkcyan', 'orange']
time_colors = ['cornflowerblue', 'aqua', 'mediumturquoise', 'coral']

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Bars for supply and time to sell with custom colors
bars1 = ax.bar(x - bar_width/2, df['Supply Change (%)'], width=bar_width, label='Supply Change (%)', color=supply_colors)
bars2 = ax.bar(x + bar_width/2, df['Time to Sell Change (%)'], width=bar_width, label='Time to Sell Change (%)', color=time_colors)

# Add value labels above bars
for bar in bars1:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height + 1, f'+{height}%', ha='center', va='bottom', fontsize=9)

for bar in bars2:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height + 1, f'+{height}%', ha='center', va='bottom', fontsize=9)

# Customize the plot
ax.set_xticks(x)
ax.set_xticklabels(df['Property Type'], fontsize=10, fontweight='bold')
ax.set_ylabel('PERCENTAGE INCREASE', fontsize=10, fontweight='bold')
ax.set_title('CHANGE IN SUPPLY AND TIME ON MARKET BY PROPERTY TYPE \n BRISTOL july 2024 - JULY 2025)', fontsize=14, fontweight='bold')
ax.legend()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Data for number of properties advertised for sale in Bristol
data = {
    'Property Type': ['Detached', 'Semi', 'Terraced', 'Flat'],
    'July 2024': [446, 801, 532, 1037],
    'July 2025': [525, 1032, 810, 1345]
}

# Create DataFrame
df = pd.DataFrame(data)

# Calculate percentage change
df['% Change'] = ((df['July 2025'] - df['July 2024']) / df['July 2024'] * 100).round(0).astype(int)

# Plotting grouped bar chart
x = np.arange(len(df))
bar_width = 0.35

fig, ax = plt.subplots(figsize=(9, 5))

# Define custom colors for each property type
colors_jan = ['skyblue', 'lightcyan', 'darkcyan', 'orange']
colors_jul = ['cornflowerblue', 'aqua', 'mediumturquoise', 'coral']

# Plot bars for January 2007 and July 2025
bars1 = ax.bar(x - bar_width/2, df['July 2024'], width=bar_width, color=colors_jan)
bars2 = ax.bar(x + bar_width/2, df['July 2025'], width=bar_width, color=colors_jul)

# Add percentage change labels above July 2025 bars only
for i, (x_pos, pct) in enumerate(zip(x, df['% Change'])):
    ax.text(x_pos + bar_width/2, df['July 2025'][i] + 20, f'+{pct}%\nProperties', ha='center', fontsize=9)

# Customizing the plot
ax.set_xticks(x)
ax.set_xticklabels(df['Property Type'], fontweight='bold')
ax.set_ylabel('NUMBER OF PROPERTIES ADVERTISED FOR SALE', fontweight='bold')
ax.set_title('CHANGE IN NUMBER OF PROPERTIES ADVERTISED FOR SALE \nIN BRISTOL BY PROPERTY TYPE (JULY 2024 - JULY 2025)', fontsize=12, fontweight='bold')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(False)

plt.tight_layout()
plt.show()




```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Data for average time on market in January 2007 and July 2025
data = {
    'Property Type': ['Detached', 'Semi', 'Terraced', 'Flat'],
    'July 2024': [100, 86, 88, 135],
    'July 2025': [111, 93, 101, 140]
}

df = pd.DataFrame(data)

# Calculate percentage change
df['% Change'] = ((df['July 2025'] - df['July 2024']) / df['July 2024'] * 100).round(1)

# Plotting grouped bar chart
x = np.arange(len(df))
bar_width = 0.35

fig, ax = plt.subplots(figsize=(8, 5))

# Define custom colors for each property type
colors_jan = ['skyblue', 'lightcyan', 'darkcyan', 'orange']
colors_jul = ['cornflowerblue', 'aqua', 'mediumturquoise', 'coral']

# Plot bars for January 2007 and July 2025
bars1 = ax.bar(x - bar_width/2, df['July 2024'], width=bar_width, color=colors_jan)
bars2 = ax.bar(x + bar_width/2, df['July 2025'], width=bar_width, color=colors_jul)

# Add percentage change labels above July 2025 bars
for i, (x_pos, pct) in enumerate(zip(x, df['% Change'])):
    ax.text(x_pos + bar_width/2, df['July 2025'][i] + 2, f'+{pct}%\n increase', ha='center', fontsize=9)

# Customizing the plot
ax.set_xticks(x)
ax.set_xticklabels(df['Property Type'], fontweight='bold')
ax.set_ylabel('AVERAGE NUMBER OF DAYS ON MARKET', fontweight='bold')
ax.set_title('CHANGE IN AVERAGE TIME ON MARKET \nBY PROPERTY TYPE (JULY 2024 - JULY 2025)', fontsize=12, fontweight='bold')
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.grid(False)

plt.tight_layout()
plt.show()

```
