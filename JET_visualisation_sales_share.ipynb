```python
# Count frequency of orders from each country
country_count = df['Country'].value_counts()
print(country_count)
```

```python


# Let's display these results in a very basic bar chart
plt.figure(figsize=(10,6))
country_count.plot(kind='bar', color='red')

# Add titles to the chart
plt.title('Sales Distribution by Country', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Country', fontsize=12, fontweight='bold', labelpad=20)
plt.ylabel('Frequency', fontsize=12, fontweight='bold', labelpad=20)

# Customise and display
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
'''
It is clear to see that despite sales in the USA being the highest, the sales per capita are lower, especially compared to the UK.
We could then investigate the total amount of sales in each country - which countries actually spend the most?
Before we do that, let's look at other ways of visualising this information (both good and bad!)
'''
```

```python
# Let's do this bar chart again with another colour style:

plt.figure(figsize=(10, 6))

# List of sequential blue shades -  we have seen from the previous bar chart that sales in Australia and Canada are equal,
# so I'm going to make these two countries the same shade
colors = ['#084594', '#2171b5', '#4292c6', '#c6dbef', '#c6dbef']

# Plot the bar chart with sequential colours
country_count.plot(kind='bar', color=colors[:len(country_count)])

# Add titles to the chart
plt.title('Sales Distribution by Country', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Country', fontsize=12, fontweight='bold', labelpad=20)
plt.ylabel('Frequency', fontsize=12, fontweight='bold', labelpad=20)

# Customise and display
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Let's make a pie chart with the same colours! A pie chart is useful because it shows
# the percentage of total sales for each country.

plt.figure(figsize=(8, 8))

country_count.plot(kind='pie',
                   autopct='%1.1f%%',  # Adding percentages
                   startangle=90,  # Rotate start angle
                   colors = ['#084594', '#2171b5', '#4292c6', '#c6dbef', '#c6dbef'],
                   wedgeprops={'edgecolor': 'white', 'linewidth': 2})  # Adding dividing lines)


# Add title to the chart
plt.title('Sales Distribution by Country', fontsize=18, fontweight='bold', pad=20)

# Display the plot
plt.ylabel('')  # Removes default y-axis label
plt.tight_layout()
plt.show()
```

```python
# Let's change this slightly so the labels and percentages are inside the pie, the font is a bit bigger
# and let's change the colour scheme to something a bit more colourblind friendly.

sizes = country_count.values  # Sales data (values)
labels = country_count.index  # Country names (index)
colors = ['#332288', '#117733',	'#44aa99', '#88ccee', '#ddcc77']
# Custom colourblind friendly shades using Color-Blind Distinctive Part 1 by Paul Tol Color Palette

# Create chart!
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    sizes,
    labels=None,  # Removing default labels that usually display outside of the chart
    autopct='',  # Remove default percentage handling
    startangle=90,
    colors=colors,
    wedgeprops={'edgecolor': 'white', 'linewidth': 2}
)

# Adding in the combined labels (Country + Percentage) inside of each wedge
for i, wedge in enumerate(wedges):
    angle = (wedge.theta1 + wedge.theta2) / 2  # Locate midpoint of the wedge
    x = 0.6 * np.cos(np.radians(angle))      # Adjust radius for label placement
    y = 0.6 * np.sin(np.radians(angle))
    percentage = sizes[i] / sum(sizes) * 100  # Calculate percentage
    label = f'{labels[i]}\n{percentage:.1f}%'  # Combine country and percentage
    plt.text(x, y, label, ha='center', va='center', fontsize=12, fontweight='bold', color='white')

# Add Title
plt.title('Sales Distribution by Country \n (Colourblind Distinctive)', fontsize=18, fontweight='bold')

# Display the Chart
plt.tight_layout()
plt.show()


```

```python
# Now let's make a stacked line chart. This is not really appropriate for this kind of sales data as
# it implies a cumulative relationship between the countries. It overcomplicates simple data.


plt.figure(figsize=(10, 6))
plt.stackplot(country_count.index, country_count.values, colors=['skyblue'], alpha=0.6)
plt.title('Stacked Line Chart (Misleading)', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Country', fontsize=12, fontweight='bold', labelpad=20)
plt.ylabel('Frequency', fontsize=12, fontweight='bold', labelpad=20)
plt.tight_layout()
plt.show()
```

```python
# Now I want to look at the overall sales split between men and women

# Count gender split
gender_count = df['Gender'].value_counts()
print(gender_count)
```

```python

plt.figure(figsize=(8, 8))

# Data
labels = ['Male', 'Female']
sizes = [(186969/300710)*100, (113741/300710)*100]
colors = ['#4292c6', '#cc79a7']


# Create donut chart
plt.figure(figsize=(8, 8))
wedges, texts, autotexts = plt.pie(
    sizes,
    labels=None,  # Remove outside labels
    autopct='%1.1f%%',  # Add percentages inside
    startangle=90,
    colors=colors,
    wedgeprops={'edgecolor': 'white', 'linewidth': 2}
)
# Customise Percentages (autotexts)
for autotext in autotexts:
    autotext.set_color('white')  # Set text color to white
    autotext.set_fontsize(10)    # Adjust font size
    autotext.set_fontweight('bold')  # Make text bold

# Add labels into the wedges
for i, wedge in enumerate(wedges):
    angle = (wedge.theta1 + wedge.theta2) / 2  # Midpoint of each wedge
    x = 0.6 * np.cos(np.radians(angle))       # Adjust radius for labels
    y = 0.8 * np.sin(np.radians(angle))
    plt.text(x, y, labels[i], ha='center', va='center', fontsize=10, fontweight='bold', color='white')

# Add donut hole
center_circle = plt.Circle((0, 0), 0.4, color='white')
plt.gca().add_artist(center_circle)


plt.title('Gender Distribution in Sales', fontsize=18, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# The youngest customer age is 18 and the oldest is 70. Let's split the ages now between gender:

age_gender_data = df[['Age', 'Gender']]

# Group by 'Age' and 'Gender' and count the occurrences
age_gender_counts = age_gender_data.groupby(['Age', 'Gender']).size().unstack(fill_value=0)
print(age_gender_counts)
```

```python
# Now let's plot this on a graph. We'll stick to pink for girls and blue for boys:

# Plotting the lines for Male and Female
plt.figure(figsize=(12, 6))
plt.plot(age_gender_counts.index, age_gender_counts['Male'], marker='o', color='#4292c6', label='Male', linewidth=2)
plt.plot(age_gender_counts.index, age_gender_counts['Female'], marker='o', color='#cc79a7', label='Female', linewidth=2)

# Adding titles and labels
plt.title('Worldwide Sales Distribution \n by Age and Gender', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Age', fontsize=14, fontweight='bold', labelpad=20)
plt.ylabel('Number of Sales', fontsize=14, fontweight='bold', labelpad=20)
plt.xticks(ticks=range(int(age_gender_counts.index.min()), int(age_gender_counts.index.max()) + 1, 2))  # X-ticks every 2 years

plt.legend(title='Gender', fontsize=12)

# Adding grid and layout
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.tight_layout()

# Display the graph
plt.show()
```

```python
# The above graph shows quite clearly that the number of sales peaks and falls at the same ages for both genders.
# Highest ages for sales for Males/Females - 26 / 20
# Second highest 20 / 46

# The main peaks are early - mid 20s and then mid 40s
# What items are selling to these age groups that don't seem to appeal to people over 50?
```

```python

# Filter the df for UK again
uk_data = df[df['Country'] == 'UK']

# Then filter for ages under 30
uk_data_under30 = uk_data[uk_data['Age'] < 30]

# Group by Gender and Income, and count occurrences
uk_income_gender_count_under30 = uk_data_under30.groupby(['Gender', 'Income']).size().unstack(fill_value=0)

print(uk_income_gender_count_under30)


```

```python

uk_income_gender_count_under30.plot( kind='bar', figsize=(12, 6),
                                    color=['#66CC66','#FF6666', '#FFC966'], # Assign colours to income brackets
                                    edgecolor='black' ) # Add titles and labels

plt.title('Income Bracket by Gender \n Under 30s - UK', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Gender', fontsize=14, fontweight='bold', labelpad=20)
plt.ylabel('Number of Sales', fontsize=14, fontweight='bold', labelpad=20)
plt.legend(title='Income Bracket', fontsize=12)
plt.xticks(rotation=0) # Keep gender labels horizontal

# Add grid for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

```

```python
uk_income_gender_count_under30.plot(
    kind='bar',
    stacked=True,
    figsize=(12, 6),
    color=['#FF6666', '#FFC966', '#66CC66'],
    edgecolor='black'
)
plt.title('Income Bracket by Gender \n Under 30s - UK', fontsize=18, fontweight='bold', pad=20)
plt.xlabel('Gender', fontsize=14, fontweight='bold', labelpad=20)
plt.ylabel('Number of Sales', fontsize=14, fontweight='bold', labelpad=20)
plt.legend(title='Income Bracket', fontsize=12)
plt.xticks(rotation=0)
plt.tight_layout()
```

```python
# An unhelpful way to view the same data.
# Bubble Chart - we can see from the result below that the medium income bracket is the
# largest, it's just not a very clear display, especially with comparison to the bar chart above.


genders = uk_income_gender_count_under30.index  # Gender (Male, Female)
income_brackets = uk_income_gender_count_under30.columns  # Income Brackets (Low, Medium, High)

# Prepare data for the bubble chart
x = []  # X-axis values (Gender as numeric)
y = []  # Y-axis values (Income Brackets as numeric)
bubble_sizes = []  # Bubble sizes (Count of customers)

for i, gender in enumerate(genders):
    for j, income in enumerate(income_brackets):
        x.append(i)  # Gender (0 = Male, 1 = Female)
        y.append(j)  # Income Bracket (0 = Low, 1 = Medium, 2 = High)
        bubble_sizes.append(uk_income_gender_count_under30.loc[gender, income] * 10)  # Scale bubble sizes

# Bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=bubble_sizes, color='#66CC66', alpha=0.6, edgecolor='black')

# Customising the axes
plt.xticks([0, 1], labels=genders, fontsize=12)  # Gender labels
plt.yticks([0, 1, 2], labels=income_brackets, fontsize=12)  # Income labels
plt.xlabel('Gender', fontsize=14, fontweight='bold')
plt.ylabel('Income Bracket', fontsize=14, fontweight='bold')

# Add Title
plt.title('Income Bracket by Gender \n Under 30s - UK', fontsize=18, fontweight='bold', pad=20)

# Add grid and layout
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
